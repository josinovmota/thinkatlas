## Aqui está uma clara representação de soma de vetores. Ao somar cada vetor do dinossauro por (-1, -2.5) deslocamos ele no plano

```python
dino_vectors = [(6, 4), (3, 1), (1, 2), (-1, 5), (-2, 5), (-3, 4), (-4, 4), (-5, 3), (-5, 2), (-2, 2), (-5, 1), (-3, 0), (-2, 1), (-1, 0), (0, -3), (-1, -4), (1, -4), (2, -3), (1, -2), (3, -1), (5, 1)]

dino_vectors2 = [numpy.add((-1,-2.5), v) for v in dino_vectors]

draw(Points(*dino_vectors, color=blue),
     Polygon(*dino_vectors, color=blue),
     Points(*dino_vectors2, color=red),
     Polygon(*dino_vectors2, color=red))
```

## Conseguimos notar que a subtração de vetores, nada mais é do que a distância da `tip` de um vetor para a `tip` do outro. Nesse exemplo temos tanto ela partindo da `origem`, como ela representada `tip-to-tip`

```python
draw(Points((-1,3), (2,2)),
     Arrow((-1,3), (0,0), color=blue),
     Arrow((2,2), (0,0), color=blue),
     # Subtração vetor1 e vetor2
     Points((-1-2, 3-2)),
     # Seta partindo da origem
     Arrow((-3,1), (0,0)),
     # Seta partindo da tip do vetor2 e parando na tip do vetor1
     Arrow((-1,3), (2, 2)))
```

## Exercício 2.7

```python
def add(*vectors):
     sum_of_x = sum([v[0] for v in vectors])
     sum_of_y = sum([v[1] for v in vectors])

     return((sum_of_x, sum_of_y))

print(add((1,5), (2,5), (3,5)))
```

## Exercício 2.8

```python
def translate(translation, vectors):
     return([add(translation, v) for v in vectors])

print(translate((1,1), [(0,0), (0,1), (-3,-3)]))
```
