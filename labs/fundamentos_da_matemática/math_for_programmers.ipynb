## Exercício 2.20

A questão pede para provar algébricamente que um vetor tem o mesmo length que oposto dele mesmo.

Ou seja:

(a, a) mesmo length que (-a, -a) ou seja que:

$\sqrt{a^2 + a^2} = \sqrt{(-a)^2 + (-a)^2}$

Bom, a única diferença notável dos dois é que um tem seu v[0] e v[1] positivos e o outro, negativos

Mas, sabemos que todo `número negativo` **multiplicado** por outro `número negativo` da um `número positivo`, ou seja
se utilizarmos o vetor (1,1) e seu oposto, teremos:

$\sqrt{1 + 1} = \sqrt{1 + 1}$

Tornando assim, a igualdade correta

## Exercício 2.25

```python
dino_vectors = [(6, 4), (3, 1), (1, 2), (-1, 5), (-2, 5), (-3, 4), (-4, 4), (-5, 3), (-5, 2), (-2, 2), (-5, 1), (-3, 0), (-2, 1), (-1, 0), (0, -3), (-1, -4), (1, -4), (2, -3), (1, -2), (3, -1), (5, 1)]

def distance(vector1, vector2):
    vectors_subtraction = subtract(vector1, vector2)

    vectors_distance = length(vectors_subtraction)

    return vectors_distance

# v1, v2, v3, v4

def perimeter(*vectors):

    vectors_sum = 0

    for vector_index in range(len(*vectors) - 1):

        vectors_sum += distance(vectors[0][vector_index], vectors[0][vector_index + 1])

    vectors_sum += distance(vectors[0][0], vectors[0][-1])

    return vectors_sum

print(perimeter(dino_vectors))
```
